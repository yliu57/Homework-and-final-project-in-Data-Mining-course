```python
import pandas as pd
import numpy as np

data = pd.read_csv('./train.csv', header = 0)
data
```

```python
race_AA = data[data['race']=='African-American'].index
race_AA
```

```python
race_CA = data[data['race']=='Caucasian'].index
race_CA
```

```python
data_label = data[['label']]
# data_label = data_label.astype(int)
data_label
```

```python
print(data_label.iat[5047,0])
```

```python
data_features = data.drop(columns='label')
data_features
```

```python
data_features_text = data_features[['c_charge_desc']]
# data_features_text = data_features_text.to_numpy()
# data_features_text = pd.DataFrame(data_features_text)
data_features_text
```

```python
data_features_ohe = data_features[['sex','age_cat','c_charge_degree']]
# data_features_ohe = data_features_ohe.to_numpy()
# data_features_ohe = pd.DataFrame(data_features_ohe)
data_features_ohe
```

```python
data_features_num = data_features[['age','juv_fel_count','juv_misd_count','juv_other_count','priors_count']]
# data_features_num = data_features_num.to_numpy()
# data_features_num = pd.DataFrame(data_features_num)
data_features_num
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(dtype=int)# dtype=int
# data_features = data_features
enc.fit_transform(data_features_ohe)
print("enc.categories_:",enc.categories_)
enc_data_features = enc.transform(data_features_ohe).toarray()
# print('enc_data_features:', enc_data_features)
enc_data_features = pd.DataFrame(enc_data_features) 
print('enc_data_features:', enc_data_features)
```

```python
# data_features_text = data_features_text.to_numpy()
data_features_text
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(dtype=int)# dtype=int
enc.fit(data_features_text)
print("enc.categories_:",enc.categories_)
enc_data_features_text = enc.transform(data_features_text).toarray()
# print('enc_data_features:', enc_data_features)
enc_data_features_text = pd.DataFrame(enc_data_features_text) 
print('data_features_text:', enc_data_features_text)

# enc_predict_features_text = enc.transform(predict_features_text).toarray()
# # print('enc_data_features:', enc_data_features)
# enc_predict_features_text = pd.DataFrame(enc_predict_features_text) 
# print('predict_features_text:', enc_predict_features_text)
```

```python
enc_data_features = np.append(enc_data_features, enc_data_features_text, axis=1)
enc_data_features = np.append(enc_data_features, data_features_num, axis=1)
enc_data_features = pd.DataFrame(enc_data_features)
enc_data_features
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_predict
# parameters = { 'max_depth':tune_array_1_50, 'min_samples_split':[2,3,4,5], 'min_samples_leaf':[1] }
decision_tree = DecisionTreeClassifier() # random state=0?

# Get the final result
# decision_tree = decision_tree.fit(enc_data_features, data_label_5000)
# label_predict = decision_tree.predict(enc_data_features_526)
# print("length of label_predict:",len(label_predict))
prediction = cross_val_predict(decision_tree, enc_data_features, data_label, cv=5)
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
print("f1 score average = None :", f1_score(data_label, prediction, average=None))
print("f1 score average = macro :", f1_score(data_label, prediction, average='macro'))
print("accuracy_score:", accuracy_score(data_label, prediction))
print("confusion_matrix:", confusion_matrix(data_label, prediction))

# # Grid Search hyperparameter tuning
# tune_array_1_20 = list(range(1,21))
# tune_array_2_5 = list(range(2,6))
# tune_array_1_10 = list(range(1,11))
# parameters = { 'max_depth':tune_array_1_20, 'min_samples_leaf':tune_array_1_30,'min_samples_split':tune_array_2_5 }
# clf = GridSearchCV(estimator = decision_tree, param_grid = parameters, scoring = 'accuracy', cv = 3, return_train_score = True, n_jobs=-1)
# clf.fit(enc_data_features_5000, data_label_5000)
# print("best parameters are ", clf.best_params_)
# label_predict = clf.predict(enc_data_features_526)
# print("length of label_predict:",len(label_predict))
# from sklearn.metrics import accuracy_score
# from sklearn.metrics import confusion_matrix
# print("accuracy_score:", accuracy_score(data_label_526, label_predict))
# print("confusion_matrix:", confusion_matrix(data_label_526, label_predict))

# from sklearn.ensemble import RandomForestClassifier
# decision_tree = RandomForestClassifier(random_state=0)
# decision_tree = decision_tree.fit(enc_data_features_5000, data_label_5000)
# label_predict = decision_tree.predict(enc_data_features_526)
# print("length of label_predict:",len(label_predict))
# from sklearn.metrics import f1_score
# from sklearn.metrics import accuracy_score
# from sklearn.metrics import confusion_matrix
# print("f1 score average = None :", f1_score(data_label_526, label_predict, average=None))
# print("f1 score average = macro :", f1_score(data_label_526, label_predict, average='macro'))
# print("accuracy_score:", accuracy_score(data_label_526, label_predict))
# print("confusion_matrix:", confusion_matrix(data_label_526, label_predict))
```
